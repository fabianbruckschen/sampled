# Visualize the runtime analysis as a line graph

## Preparations

```python
# modules
import matplotlib
import matplotlib.pyplot as plt  # for saving plots
import pandas as pd
```

```python
# plotting options
%matplotlib inline
```

```python
# load data
df = pd.read_csv('data/timer.csv', names=['level', 'timestamp', 'n_secrets', 'seconds'])
```

### Splitting Data

```python
total = df[['n_secrets', 'seconds']].groupby('n_secrets').agg('sum')
total['level'] = 'Total time'
```

```python
level0 = df[df.level=='0: Data Creation and Encryption']
level1 = df[df.level=='1: Secret Sharing']
level2 = df[df.level=='2: Aggregated Value Broadcasting']
```

## Plotting

```python
xlabels = ['1e2', '1e3', '1e4', '1e5', '1e6']
```

```python
pcolor = '#34DCB0'
matplotlib.rc('axes', edgecolor=pcolor, facecolor='none', 
              labelcolor=pcolor, linewidth=3)
matplotlib.rc('lines', lw=3)
```

```python
ax = plt.subplot(111)
ax.plot(xlabels, level0.seconds, color=pcolor)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel('Number of secrets', fontsize=14)
ax.set_ylabel('Seconds passed', fontsize=14)
ax.tick_params(axis='both', colors=pcolor, labelsize=12)
ax.set_title('Level 0: Data Creation and Encryption\n', color=pcolor, fontsize=18)
```

```python
ax = plt.subplot(111)
ax.plot(xlabels, level1.seconds, color=pcolor)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel('Number of secrets', fontsize=14)
ax.set_ylabel('Seconds passed', fontsize=14)
ax.tick_params(axis='both', colors=pcolor, labelsize=12)
ax.set_title('Level 1: Secret Sharing\n', color=pcolor, fontsize=18)
```

```python
ax = plt.subplot(111)
ax.plot(xlabels, level2.seconds, color=pcolor)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel('Number of secrets', fontsize=14)
ax.set_ylabel('Seconds passed', fontsize=14)
ax.tick_params(axis='both', colors=pcolor, labelsize=12)
ax.set_title('Level 2: Aggregated Value Broadcasting\n', color=pcolor, fontsize=18)
```
